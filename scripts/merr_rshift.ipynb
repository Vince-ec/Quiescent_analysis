```python
import numpy as np
from glob import glob
from spec_id import Galaxy_ids
import pandas as pd
from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt
import os
from shutil import copyfile
import seaborn as sea
from astropy.table import Table
sea.set(style='white')
sea.set(style='ticks')
sea.set_style({"xtick.direction": "in","ytick.direction": "in"})
cmap = sea.cubehelix_palette(12, start=2, rot=.2, dark=0, light=1.1, as_cmap=True)
```

```python
flistl = glob('../mcerr/*s39170*')
flistm = glob('../mcerr/*n21156*')
flisth = glob('../mcerr/*s40862*')
flistuh = glob('../mcerr/*s44042*')
```

```python
mfitsl,mfitsm,mfitsh,mfitsuh = np.array([[],[],[],[]])
afitsl,afitsm,afitsh,afitsuh = np.array([[],[],[],[]])

for i in range(len(flistl)):
    m, a = np.load(flistl[i])
    mfitsl = np.append(mfitsl,m)
    afitsl = np.append(afitsl,a)
    
    m, a = np.load(flistm[i])
    mfitsm = np.append(mfitsm,m)
    afitsm = np.append(afitsm,a)
    
    m, a = np.load(flisth[i])
    mfitsh = np.append(mfitsh,m)
    afitsh = np.append(afitsh,a)
    
    m, a = np.load(flistuh[i])
    mfitsuh = np.append(mfitsuh,m)
    afitsuh = np.append(afitsuh,a)
```

```python
%matplotlib inline

plt.figure(figsize=[8,8])
sea.kdeplot(mfitsl,afitsl)
plt.scatter(0.015,3.2)
plt.axis([0,0.03,0,6])
plt.show()
plt.close()

plt.figure(figsize=[8,8])
sea.kdeplot(mfitsm,afitsm)
plt.scatter(0.015,3.2)
plt.axis([0,0.03,0,6])
plt.show()
plt.close()

plt.figure(figsize=[8,8])
sea.kdeplot(mfitsh,afitsh)
plt.scatter(0.015,3.2)
plt.axis([0,0.03,0,6])
plt.show()
plt.close()

plt.figure(figsize=[8,8])
sea.kdeplot(mfitsuh,afitsuh)
plt.scatter(0.015,3.2)
plt.axis([0,0.03,0,6])
plt.show()
plt.close()
```

```python
plt.figure(figsize=[8,8])
sea.distplot(mfitsl)
sea.distplot(mfitsm)
sea.distplot(mfitsh)
sea.distplot(mfitsuh)
plt.show()
plt.close()
```
